```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

### Esericizio 02.1

Si vuole calcolare il seguente integrale
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$
attraverso due tecniche:

1. campionando la distribuzione uniforme nel range $[0,1]$
2. usando il metodo "importance sampling"

Tale tecniche derivano dall'osservazione che l'integranda può essere decomposta in due funzioni di cui una può essere considerata la distribuziione di probabilità.
Nel primo caso $\rho(x)$=1, quindi non bisogna aggiungere nulla.
Nel secondo si vuole descrivere la densità di probabilità con una funzione simile all'altro pezzo di integranda; è proprio questo il nocciolo di questa tecnica. La distribuzione scelta $\rho(x)=-2x-2$. Per campionarla basta calcolare la funzione cumulativa ed invertirla.

Le figure seguenti mostrano, da sinistra verso destra, l'applicazione del primo e del secondo metodo.
I valori ottenuti sono compatibili con il valore teorico.

```python
fig, (f1, f2) = plt.subplots(1,2,figsize=(13,3.5))

x, f, error = np.loadtxt("./Es2.1./Statistica0.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
f1.errorbar(x,f,yerr=error,color="r")
f1.set_ylabel("$Integrale$")
f1.set_xlabel("$N$")
f1.set_title("Calcolo dell'integrale: distribuzione uniforme")
f1.grid(True)

x, f, error = np.loadtxt("./Es2.1./Statistica1.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error, color="orange")
f2.set_ylabel("$Integrale$")
f2.set_xlabel("$N$")
f2.set_title("Calcolo dell'ntegrale: Importance sampling")
f2.grid(True)
plt.show()
```

### Esercizio 02.2

Un random walk è un percorso nel quale ogni passo è casuale. La dinamica è descritta dall'equazione di diffusione.

In questo esercizio si vuole simulare un random walk dove ogni passo può avvenire nelle tre direzioni ed ha ampiezza unitaria. Graficando $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ contro il numero di steps si ottiene la figura sottostante a sinistra. 

Si può ripetere il calcolo facendo tendere l'ampiezza del passo a zero. In questo modo, il random walk si sviluppa in tre dimensioni senza vincoli nella dinamica. 
Si veda il grafico in basso a destra.

Dai grafici si nota che $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ è una funzione proporzionale alla radice di N, dove N indica di quanti passi mi sono allontanata dalla partenza.
Il fit mostra tale andamento.

Gli errori sono piccoli rispetto alla scala con la quale varia la funzione in esame. Per tale motivo ho voluto graficarli facendo un fit. Qualitativamente anche tale funzione è proporzionale alla radice del passo.

```python
def f(x,b,a):
    return a+b*pow(x,0.5)

fig, (f1, f2) = plt.subplots(1,2,figsize=(14,3))

x, y, error = np.loadtxt("./Es2.2.a./Statistica.dat", usecols=(0,1,2), delimiter=' ',  unpack='true')
f1.errorbar(x,y,yerr=error,color="r")
f1.set_title("Random walk discreto")
popt, pcov = curve_fit(f, x, y)
print(popt)
f1.plot(x, f(x, *popt),color="b")
f1.grid(True)

f2.plot(x,error)
f2.set_title("Errori: random walk discreto")
popt, pcov = curve_fit(f, x, error)
print(popt)
f2.plot(x, f(x, *popt),color="b")
f2.grid(True)
plt.show()

fig, (f1, f2) = plt.subplots(1,2,figsize=(14,3))

x, y, error = np.loadtxt("./Es2.2.b./Statistica.dat", usecols=(0,1,2), delimiter=' ',  unpack='true')
f1.errorbar(x,y,yerr=error,color="r")
f1.set_title("Random walk discreto")
popt, pcov = curve_fit(f, x, y)
print(popt)
f1.plot(x, f(x, *popt),color="b")
f1.grid(True)

f2.plot(x,error)
f2.set_title("Errori: random walk discreto")
popt, pcov = curve_fit(f, x, error)
print(popt)
f2.plot(x, f(x, *popt),color="b")
f2.grid(True)
plt.show()
```
