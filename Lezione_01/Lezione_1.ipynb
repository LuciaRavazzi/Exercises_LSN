### Esercizio 1.1


È stato fornito un generatore di numeri casuali tra [0,1) di cui si vuole verificare che 
         $\langle r \rangle = \int_0^1 r dr = 1/2$. 

Inizialmente sono stati generati $M= 10^7$ numeri casuali i quali sono stati suddivisi in $N=100$ blocchi. Per ogni blocco è stata calcolata la media. L'errore di ogni blocco viene calcolato a partire dalle somme progressive dalle quali si ricava la deviazione standard.

Il grafico sottostante mostra la convergenza dei valori ottenuti rispetto al valore teorico.

```python
x, f, error = np.loadtxt("./Es1.1.a./risultati.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f-0.5,yerr=error)
plt.xlabel('N')
plt.ylabel('$<r>-1/2$')
plt.title("Test generatore numeri casuali",fontsize=14)
plt.grid(True)
plt.show()
```

Si vuole verificare che la deviazione standard della media $\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12$. 

Il procedimento utilizzato è identico a quello utilizzato nell'esercizio precedente.

```python
x, f, error = np.loadtxt("./Es1.1.b./risultati.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f-1/12,yerr=error)
plt.xlabel('N')
plt.ylabel('$\sigma^2-1/12$')
plt.title("Test generatore numeri casuali")
plt.grid(True)
plt.show()
```

Si vuole ora verificare l'uniformità del generatore fornito attraverso il test del $\chi^2$.
Si divide l'intervallo $[0,1]$ in $M=10^2$ intervalli per poi assegnare ad ogni valore estratto il corrispondente sottointervallo. Alla fine di tale procedura si calcola
$$\chi^2 = \sum_{i=1}^M \frac{\left( n_i - n/M \right)^2}{n/M}$$
dove $n_i$ è il numero di valori estratti appartenenti all'i-esimo sottoinvervallo, $n$ è il numero totale di lanci (pari a $10^4$). In media, $(n_i - n/M)^2 \simeq n/M$ e quindi $\chi^2 \simeq 100$, cioè il numero di sottointervalli. Ottenendo tale risultato si può affermare che il generatore fornisce numeri casuali distributi in modo uniforme nell'intervallo.

Dalla figura sottostante si osserva che ci sono forti fluttuazioni in un intorno del valore voluto. Si può concludere che il generatore fornito è abbastanza buono.

```python
x, y= np.loadtxt("./Es1.1.c./test.dat", usecols=(0,1), delimiter=' ', unpack='true')
plt.plot(x,y,marker=".")
plt.xlabel('N',fontsize=14)
plt.ylabel('$X^2$',fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=11.5)
plt.grid(True)
plt.show()
```

### Esercizio 1.2

In questo esercizio si vuole verificare il teorema del limite centrale campionando diverse distribuzioni di probabilità:

-uniforme $p(x)=1$ con $x\in [0;1)$

-esponenziale
$p(x) = \lambda \exp(-\lambda x)$ con $x\in [0;+\infty]$

-lorentziana 
$p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}$, $x\in [-\infty;+\infty]$

Per il campionamento delle distribuzioni, si è trovata l'inversa della funzione cumulativa. <br>
Per un numero molto grande di lanci, si osserva che gli istogrammi delle prime due distribuzioni tendono alla tipica forma gaussiana confermando la validità del teorema del limite centrale. Il confronto con la funzione gaussiana è realizzato dal fit.  

Tali istogrammi sono stati ottenuti generando $N$ numeri ($N$ è scritto nella legenda) per poi inserire 
$$
S_n= \frac{1}{N}  \sum\limits_{n=1}^N  x_n
$$
nel corretto bin.

Per quanto riguarda la **distribuzione uniforme**, si osserva che la media dell'istogramma già nella seconda figura combacia con quella predetta dal teorema del limite centrale (0.5).

```python
import math
fig, (f1, f2, f3, f4) = plt.subplots(1,4,figsize=(20,5))

x,y = np.loadtxt("./Es1.2./uniforme1.dat", usecols=(0,1), unpack='true')
f1.plot(x,y,label='N=1',marker=".",color="crimson")
leg = f1.legend(fontsize=15);
f1.tick_params(axis='both', which='major', labelsize=14.5)
f1.grid(True)

y= np.loadtxt("./Es1.2./uniforme2.dat", usecols=(1), unpack='true')
f2.plot(x,y, marker=".",label='N=2',color="salmon")
leg = f2.legend(fontsize=15);
f2.tick_params(axis='both', which='major', labelsize=14.5)
f2.grid(True)

y= np.loadtxt("./Es1.2./uniforme10.dat", usecols=(1), unpack='true')
f3.plot(x,y,marker=".",label='N=10',color="orangered")
leg = f3.legend(fontsize=15);
f3.tick_params(axis='both', which='major', labelsize=14.5)
f3.grid(True)

def func(x, a, mu, sigma):
     return (a/(pow(2*math.pi,2)*sigma))* np.exp(-((x-mu)**2)/(2*sigma)) 

y= np.loadtxt("./Es1.2./uniforme100.dat", usecols=(1), unpack='true')
f4.plot(x,y,marker=".",label='N=100',color="red")
leg = f4.legend(fontsize=15);
f4.tick_params(axis='both', which='major', labelsize=11.5)
f4.grid(True)
    
popt, pcov = curve_fit(func, x, y)
y_fit = func(x,popt[0],popt[1],popt[2])
print(popt)

plt.plot(x, y_fit,color="yellow")

plt.show()
```

Anche per la **distribuzione esponenziale** valgono le considerazioni di cui sopra.

Il teorema del limite centrale vale solo per quelle distribuzioni di probabilità caratterizzate dai primi due momenti non divergenti.
È noto che la distribuzione di Cauchy-Lorentz ha varianza infinita e quindi, viola tali ipotesi. 
Infatti, tale distribuzione tende ad una Cauchy-Lorentz come raffigurato. <br>
Questo fa capire che non solo la distribuzione gaussiana è un attrattore (o punto fisso) ma anche la Cauchy-Lorentz.

```python
fig, (f1, f2, f3, f4) = plt.subplots(1,4,figsize=(20,5))

x,y = np.loadtxt("./Es1.2./lorentziana1.dat", usecols=(0,1), unpack='true')
f1.plot(x,y,label='N=1',marker=".",color="lightgreen")
leg = f1.legend(fontsize=15);
f1.grid(True)

y= np.loadtxt("./Es1.2./lorentziana2.dat", usecols=(1), unpack='true')
f2.plot(x,y, marker=".",label='N=2',color="forestgreen")
leg = f2.legend(fontsize=15);
f2.grid(True)

y= np.loadtxt("./Es1.2./lorentziana10.dat", usecols=(1), unpack='true')
f3.plot(x,y,marker=".",label='N=10',color="springgreen")
leg = f3.legend(fontsize=15);
f3.grid(True)

y= np.loadtxt("./Es1.2./lorentziana100.dat", usecols=(1), unpack='true')      
f4.plot(x,y,marker=".",label='N=100',color="mediumspringgreen")
leg = f4.legend(fontsize=15);
f4.grid(True)

plt.show()
```

### Esercizio 1.3

In questo esercizio si vuole simulare l'esperimeto di Buffon il quale porta ad una stima di $\pi$.
Si suppone di gettare un ago di lunghezza L su un asse (supponendo pbc) di larghezza d e lunghezza infinita.
Dividendo in due l'asse si possono avere due possibilità dopo il lancio: l'ago si trova a destra o a sinistra di tale metà.
Se fosse verificata la prima opzione, l'ago toccherebbe l'estremo destro dell'asse solo se la somma tra il centro dell'ago e il cos(theta) risultasse positivo. Theta è un angolo estratto casualmente in [0,$\pi/2$]  e rappresenta l'angolo compreso tra la perpendicolare all'asse passante per il centro dell'ago e la metà dell'ago che punta verso il lato dell'asse d'interesse.
Se fosse invece verificata la seconda opzione, il ragionamento sarebbe analogo con l'unica differenza che al posto della somma bisogna calcolare la differenza.
Dopo aver ripetuto tale ragionamento per diverse volte, si può stimare $\pi$ con
$$\pi = \frac{2L}{Pd} = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{d N_{hit}}$$
dove $N_{thr}$ sono i tentativi, $N_{hit}$ indica il numero di volte tale per cui si è verificata l'intersezione. 
Il grafico sottostante mostra il valore ottenuto in funzione del numero dei blocchi. 
I risultati sono compatibili con il noto valore teorico.

```python
import math
print("Valore noto", math.pi)
x, f, error = np.loadtxt("./Es1.3./risultati.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error, color="forestgreen")
plt.xlabel('$Lanci$')
plt.ylabel('$\pi$')
plt.grid(True)
plt.show()
```
