### Esercizio 3.1 

La teoria di Black-Scholes propone una soluzione analitica per il calcolo dell'opzione europea.
Si assume che l'evoluzione del prezzo sia governata dal moto browniano geometrico caratterizzato da un tasso di rischio $r$ e volatilità $\sigma$.

L'opzione è la garanzia del venditore a vendere un determinato prodotto nel futuro ad un prezzo fissato. Il cliente comprando l'opzione guadagna il diritto di poter comprare tale prodotto a quel prezzo. Si ha la "call" nel caso in cui per il cliente è conveniente comprare il prodotto dal venditore poichè il prezzo è minore di quello di mercato. Se dovesse verificarsi il contrario, si parlerebbe di put. Il problema è quantificare il valore dell'opzione. La teoria di cui sopra dà tale risposta ma spesso le ipotesi non sono verificate (e.g. volatilità non costante). Di conseguenza si introducono tecniche Monte Carlo per ottenere tali risultati.

Soluzione analitica di Black-Scholes per  t<T:
- Call:
$$ C\left[ S(t),t \right] = S(t) N(d_1) -K e^{-r(T-t)} N(d_2) $$
- Put:
$$ C\left[ S(t),t \right] = S(t) \left[ N(d_1)-1\right] -K e^{-r(T-t)} \left[ N(d_2)-1\right]$$

dove 
$$d_1 = \frac{1}{\sigma\sqrt{T-t}}\left[ \ln \left( \frac{S(t)}{K} \right) + \left( r+\frac{\sigma^2}{2} (T-t) \right)\right]$$

dove $d_2 = d_1 - \sigma\sqrt{T-t}$, e dove
$$ N(x) = \int_{-\infty}^{x} dz \, e^{-\frac{z^2}{2}} = \frac{1}{2} \left[ 1+ {\rm erf} \left( \frac{x}{\sqrt{2}}\right) \right]$$
con
$${\rm erf}(x) = \frac{2}{\sqrt{\pi}}\int_{0}^{x} dz \, e^{-z^2}$$

Il prezzo evolve secondo l'equazione
$$S(t) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right]$$
dove $W(t)\sim\mathcal{N}(0,t)$.

I risultati del metodo data-blocking per la put e la call sono ottenute nel seguente modo sono graficate nelle prime due figure.

Si può eventualmente discretizzare il calcolo di $S(t)$ iterando l'equazione precedente nel seguente modo 
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    with $Z_i \sim \mathcal{N}(0,1)$
con $t_1 < t_2 < … < t_n$.

I risultati del metodo data-blocking per le due opzioni ottenute nel seguente modo sono graficate nelle ultime due figure.
In entrambi i casi si osserva un'ottima compatibilità con i risultati ottenuti dalla soluzione analitica. 

```python
fig, (f1, f2) = plt.subplots(1,2,figsize=(13,4))
x, f, error = np.loadtxt("./Es3.1./call1.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
f1.errorbar(x,f,yerr=error)
f1.set_ylabel("$Call$", fontsize=13)
f1.set_xlabel("$N$", fontsize=13)
f1.tick_params(axis='both', which='major', labelsize=12.5)

x=np.linspace(0,100,1000)
y=(x/x)*14.975790778311286
f1.plot(x,y,color="red")
f1.grid(True)

x, f, error = np.loadtxt("./Es3.1./put1.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
f2.errorbar(x,f,yerr=error)
f2.set_ylabel("$Put$", fontsize=13)
f2.set_xlabel("$N$", fontsize=13)
f2.tick_params(axis='both', which='major', labelsize=12.5)
x=np.linspace(0,100,1000)
y=(x/x)*5.4595325819072364
f2.plot(x,y,color="red")
f2.grid(True)

plt.show()
```
