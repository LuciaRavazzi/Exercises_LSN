```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, (f1, f2) = plt.subplots(1,2,figsize=(13,4))
x, f, error = np.loadtxt("./Es3.1./call1.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
f1.errorbar(x,f,yerr=error)
f1.set_ylabel("Call")
f1.set_xlabel("N")
f1.grid(True)

x, f, error = np.loadtxt("./Es3.2./put1.dat", usecols=(0,1,2), delimiter='   ', unpack='true')
f2.errorbar(x,f,yerr=error)
f2.set_ylabel("Put")
f2.set_xlabel("N")
f2.grid(True)

plt.show()
```
