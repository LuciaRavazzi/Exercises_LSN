### Esercizio 03.1 

La teoria di Black-Scholes propone una soluzione analitica per il calcolo dell'opzione europea.
Si assume che l'evoluzione del prezzo sia governata dal moto browniano geometrico caratterizzata da un tasso di rischio $r$ e volatilità $\sigma$.

L'opzione è la garanzia del venditore a vendere un determinato prodotto nel futuro ad un prezzo fissato. Il cliente comprando l'opzione guadagna il diritto di poter comprare tale prodotto a quel prezzo. Si ha la "call" nel caso in cui per il cliente è conveniente comprare il prodotto dal venditore poichè il prezzo è minore di quello di mercato. Se dovesse verificarsi il contrario, si parlerebbe di put. Il problema è quantificare il valore dell'opzione. La teoria di cui sopra dà tale risposta ma spesso le ipotesi non sono verificate (e.g. volatilità non costante). Di conseguenza si utilizzano tecniche Monte Carlo per ottenere tali risultati.

Soluzione analitica di Black-Scholes per  t<T:
- Call:
$$ C\left[ S(t),t \right] = S(t) N(d_1) -K e^{-r(T-t)} N(d_2) $$
- Put:
$$ C\left[ S(t),t \right] = S(t) \left[ N(d_1)-1\right] -K e^{-r(T-t)} \left[ N(d_2)-1\right]$$

dove 
$$d_1 = \frac{1}{\sigma\sqrt{T-t}}\left[ \ln \left( \frac{S(t)}{K} \right) + \left( r+\frac{\sigma^2}{2} (T-t) \right)\right]$$

dove $d_2 = d_1 - \sigma\sqrt{T-t}$, e dove
$$ N(x) = \int_{-\infty}^{x} dz \, e^{-\frac{z^2}{2}} = \frac{1}{2} \left[ 1+ {\rm erf} \left( \frac{x}{\sqrt{2}}\right) \right]$$
con
$${\rm erf}(x) = \frac{2}{\sqrt{\pi}}\int_{0}^{x} dz \, e^{-z^2}$$

I parametri utilizzati nella simulazione sono: 
- prezzo al tempo $t=0$: $S(0)=100$
- tempo per la scadenza: $T=1$
- prezzo del prodotto fissato: $K=100$
- tasso di rischio: $r=0.1$
- volatilità: $\sigma=0.25$

Il prezzo evolve secondo l'equazione
$$S(t) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right]$$
dove $W(t)\sim\mathcal{N}(0,t)$.

I risultati del metodo data-blocking per le due opzioni ottenute nel seguente modo sono graficate nelle prime due figure in basso.

Si può eventualmente discretizzare il calcolo di $S(t)$ iterando l'equazione precedente nel seguente modo 
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    with $Z_i \sim \mathcal{N}(0,1)$
con $t_1 < t_2 < … < t_n$.

I risultati del metodo data-blocking per le due opzioni ottenute nel seguente modo sono graficate nelle ultime due figure in basso.
In entrambi i casi si osserva un'ottima compatibilità con i risultati ottenuti dalla soluzione analitica.

### For comparison, Black-Scholes analytic solution:

```python
import scipy
from scipy.special import erf
from scipy import log, exp, sqrt

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```
