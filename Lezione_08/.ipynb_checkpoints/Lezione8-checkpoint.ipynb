### Esercizio 8.1
L'esercizio propone di risolvere un problema di ottimizzazione sfruttando i principi variazionali.

Per un sistema quantistico a più particelle, il principio di Rietz afferma che data una funzione d'onda di prova $\Psi_T$, l'energia variazionale $E_T$ è 
definita come

$$E_T =\frac{\int \Psi_T^{*}(\vec{r}_{1} ... \vec{r}_{n}) \hat{H} \Psi_T(\vec{r}_{1} ... \vec{r}_{n})d\vec{r}_{1} ... d\vec{r}_{n}}{\int |\Psi_T^{*}(\vec{r}_{1} ... \vec{r}_{n})|^2 d\vec{r}_{1} ... d\vec{r}_{n}} \ge E_0 $$ 

dove $E_0$ è l'energia dello stato fondamentale che si vuole calcolare.
Quando l'autofunzione di prova tenderà a quella dello stato fondamentale, l'energia variazionale si avvicinerà al minimo; la funzione di prova data dovrà quindi essere caratterizzata da paramentri che le permetteranno di approssimare nel modo migliore possibile l'autofunzione del ground state.

Nel caso in esame si ha solo una particella la cui dinamica si svolge in una dimensione.
È necessario osservare che l'energia può essere riscritta come 

$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$

esplicitando il fatto che la distribuzione di probabilità della particella è proprio

$$ \rho(x)=\frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} $$

mentre l'osservabile che si vuole misurare è
 
$$ E_l(x)=\frac{\hat H \Psi_T(x)}{\Psi_T(x)} \hspace{1cm} \hat H=\frac{-\hbar^2 \nabla^2}{2m} + V(x) \hspace{1cm} V(x) = x^4 - \frac{5}{2}x^2 $$ 

Dunque, nell'esercizio si deve campionare la $\rho(x)$ attraverso l'algoritmo di Metropolis (utilizzando una $T(x_{new}|x_{old})$ uniforme) per poi calcolare l'energia ad ogni passo. L'autofunzione di prova è la somma di due gaussiane
$$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$

la quale dipende da due parametri: media $\mu$ e deviazione standard $\sigma$.

I parametri $\mu$ e $\sigma$ che minimizzano l'energia sono:

```python
mu, sigma, ene = np.loadtxt("./Es8.2./Es8.2.a./dati.dat", usecols=(0,1,2), delimiter=' ', unpack='true')
s=ene.argmin()
print("Il valore minimo dell'energia: ",ene.min())
```

Per calcolare i migliori parametri ho cercato in modo grassolano quale potesse essere il range per poi approfondire la ricerca in tale sottointervallo.
La media delle due gaussiane che minimizza il valore dell'energia è $\mu=0.803 \pm 0.001$, mentre la variazione standard è pari a $\sigma=0.612\pm 0.001$.

Dal grafico sottostante si osservano le stime di $\langle {\hat H} \rangle_T$ e le incertezze statistiche come funzione del numero di blocchi utilizzando i parametri che minimizzano l'energia.

```python
x, f, error = np.loadtxt("./Es8.2./Es8.2.a./Statistica", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,f,yerr=error, color='r')
plt.title('Energia con ($\mu,\sigma$) ottimizzati')
plt.xlabel('N',fontsize=13)
plt.ylabel('$E$',fontsize=13)
plt.grid(True)
plt.show()
```

Attraverso i principi variazionale possiamo trovare l'autofunzione del ground state ma esistono anche altri metodi. Per esempio, è possibile riempire un istogramma con le posizioni campionate dalla $\rho (x)$ di cui sopra oppure sfruttare l'equazione di Schrodinger in forma matriciale per poi diagonalizzarla.
Si riportano i risultati ottenuti dalle seguenti procedure. <br>
Si osserva una buona compatibilità. L'unica discrepanza si osserva nei punti stazionari della prima e seconda funzione rispetto alla terza.

```python
mu=0.803
sigma=0.612

points=1000
x = np.linspace(-4.0,+4.0,num=points)
P=np.exp(-pow(x-mu,2)/pow(sigma,2))+np.exp(-pow(x+mu,2)/pow(sigma,2))+2.*np.exp(-(pow(x,2)+pow(mu,2))/pow(sigma,2))
norm=2*(3.14**(1/2))*sigma*(1.0+np.exp(-pow(mu,2)/pow(sigma,2)))
P=P/norm

fig, (f1, f2, f3) = plt.subplots(1,3,figsize=(20,5))

f1.plot(x,P,color="red")
f1.set_title("Densità di probabilità con (mu,sigma) ottimizzati",fontsize=14)
f1.set_ylabel('$|\Psi_T(x)|^2$',fontsize=14)
f1.set_xlabel('$x$',fontsize=14)
f1.grid(True)

x, f = np.loadtxt("./Es8.2./Es8.2.b./Istogramma.dat", usecols=(0,1), delimiter=' ', unpack='true')
f2.plot(x,f,marker=".",color="tomato")
f2.set_title("Istogramma",fontsize=14)
f2.set_ylabel('$|\Psi_T(x)|^2$',fontsize=14)
f2.set_xlabel('$x$')
f2.grid(True)


def Vpot(x):
    return (x**2 - 2.5)*x**2
hbar = 1
m = 1
a = 10
N = 1000 # number of iterations
# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)
# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)
# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)
# Plot a few things
scale = 0.3
f3.plot(x,(psi[0])**2,color="orange")
f3.set_title("Densità di probabilità: equazione di Schrodinger",fontsize=14)
f3.set_xlabel("x",fontsize=14)
f3.set_ylabel('$|\Psi_T(x)|^2$',fontsize=14)
f3.grid(True)
f3.set_xlim((-4.0,+4.0))

f1.tick_params(axis='both', which='major', labelsize=13.5)
f2.tick_params(axis='both', which='major', labelsize=13.5)
f3.tick_params(axis='both', which='major', labelsize=13.5)

plt.show()
```

### Exercise 8.2

È possibile ottenere l'autofunzione di ground state anche tramite il metodo "Path Integral Monte Carlo".
Per sistemi a temperatura nulla si utilizza il metodo PIGS, altrimenti viene usato il PIMC.
Tali metodi si basano sull'osservazione che un sistema quantistico di $N$ particelle è equivalente ad un sistema classico di $N$ polimeri lineari paticolari.

Prima di tutto, si applica il metodo PIGS a due autofunzioni di prova:

- $\Psi_T(x)=1$
- $\Psi_T(x) = \Psi_T^{\sigma,\mu}(x)$ con i parametri ottimizzati precedentemente trovati.

Le autofunzioni sembrano essere compatibili con quella ottenuta dal metodo variazionale.
Si osserva che quando le prime hanno un massimo o un minimo locale, le seconde tendono ad essere minori del minimo o maggiori del massimo. Probabilmente il PIGS non è così preciso quando l'autofunzione cambia in modo così pronunciato.

Variando la temperatura si osserva che il minimo in un intorno dell'origine è minore rispetto a quello che si osserva per T=0. Questo è dovuto al fatto che l'autofunzione ha dei contributi anche degli stati eccitati.
