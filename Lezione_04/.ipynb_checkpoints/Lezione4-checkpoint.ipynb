# <span style="color:green"> Numerical Simulation Laboratory (NSL) </span>
## <span style="color:blue">  Numerical exercises 4</span>

In questa esercitazione si vuole studiare la dinamica e calcolare i parametri termodinamici d'interesse di un set di $N$ particelle classiche in un box attraverso un algoritmo di Dinamica molecolare. <br>
Il sistema è immerso in un bagno termico la cui temperatura $T$ è fissata: l'ensemble statistico è il microcanonico. <br>
La configurazione iniziale è quella di un cristallo fcc che verrà modificata nel corso della simulazione. Le particelle così disposte non possono sovrapporsi le une con le altre (se le posizioni fossero random, non sarebbe garantita la precedente osservazione) . Per trovare una possibile disposizione reale si immergerà diverse volte il sistema nel bagno termico. <br>
L'evoluzione è regolata dall'algoritmo di Verlet dove un nuovo passo per ogni particella è definito
$$
\vec r(t+dt)=2\vec r(t)-\vec r(t-dt)+\frac{\vec f(t)}{m}dt^2
$$
$\vec f(t)$ è la forza, $m$ la massa delle particelle, $dt$ la lunghezza del passo.
Dopo aver fatto evolvere il sistema per un certo numero di passi, lo si vuole immergere nuovamente nel bagno termico: si riscalano le velocità per poi calcolare delle nuove posizioni che saranno quelle iniziali per un nuovo ciclo. <br>
Più precisamente si eseguono i seguenti punti

1. memorizzare $\vec{r}(t)$ e $\vec{r}(t-dt)$ 
2. calcolare $\vec{r}(t+dt)$ attraverso uno step dell'algoritmo di Verlet per ottenere $\vec{v}(t+dt/2)$
3. riscalo le velocità da  $\vec{v}(t)$ a $\vec{v}_s(t)$
4. uso $\vec{v}_s(t)$ per calcolare le vecchie posizioni $\vec{r}_{new}(t) = \vec{r}(t+dt) - dt \vec{v}_s$
5. uso $\vec{r}_{new}(t)$ e $\vec{r}(t+dt)$ come le nuove configurazioni iniziali

In questo modo continuo ad imporre la temperatura target al sistema in modo da avere delle velocità, dunque anche delle posizioni, più consone per il sistema che si vorrebbe studiare. <br>
Quando si inizia la simulazione si osserva che la temperatura si stabilizza su un valore che, in genere, è diverso da quello target; iterando diverse volte il meccanismo di cui sopra si raggiunge un valore ragionevolmente vicino a quello teorico.
Si può considerare tale procedimento come l'equilibrazione del sistema. In base alla fase scelta bisognerà iterare tale meccanismo un numero adeguato di volte. <br>
Alla fine, si può affermare che la configurazione ottenuta potrebbe essere una possibile reale rappresentazione iniziale del sistema.

Dopo questa prima fase di equilibrazione, si procede con il calcolo delle proprietà termodinamiche d'interesse: 
1. **Energia cinetica** per particella
$$
E_k=\frac{1}{2}(v_x^2+v_y^2+v_z^2)
$$
dove le velocità sono ottenute dall'algortimo di Verlet e la massa delle particelle è unitaria.
2. **Energia potenziale**: il potenziale d'interazione tra due particelle è il classico potenziale di Lennard-Jones se la loro distanza è minore del raggio di cut-off. 

Se $r<r_c$
$$ V(\vec r)=4 \epsilon \left [ \left (\frac{\sigma}{r}\right)^{12}- \left ( \frac{\sigma}{r} \right )^6 \right ]$$ 
altrimenti
$$ V(\vec r)=0$$ 
Tale potenziale descrive molto bene l'interazione tra atomi a shell chiuse, ovvero i gas nobili. Infatti, si simulano sistemi di Argon e Krypton.
3. **Energia totale** per particella: è semplicemente la somma dell'energia cinetica e potenziale per particella.

4. **Temperatura**: si ottiene dall'energia cinetica per particella dal principio di equipartizione dell'energia
$$
T = \frac{2}{3}\frac{E_k}{N}
$$

5. **Pressione**: dal teorema del viriale si ottiene
$$ P = \rho k_B T + \frac{1}{3V} \left\langle \sum_{i=1}^{N-1} \sum_{j\,(>i)}^N 48\epsilon \left[ \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^{12} - 
\frac{1}{2} \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right] \right\rangle $$

Per stimare tali quantità è stato utilizzato il solito metodo del data-blocking. <br>

Il grafico sottostante mostra che l'ensemble statistico è proprio quello microcanonico.


```python
fig, (f1) = plt.subplots(1,1,figsize=(10,5))
x, f = np.loadtxt("./Es4.3./Misure.dat", usecols=(0,1), delimiter=' ', unpack='true')
f1.plot(x,color="r",label="Energia cinetica")
f1.plot(f,color="y",label="Energia potenziale")
f1.plot(x+f,color="orange", label="Energia totale")
leg = plt.legend(fontsize=11);
f1.set_xlabel('$N$', fontsize=13)
f1.grid(True)
plt.show()
```

Adesso si vuole simulare un sistema di Argon e Krypton nei tre stati della materia.
I parametri scelti in unità di LJ sono:
1. fase solida: $\rho^\star = 1.1$, $T^\star = 0.8$ (raggio di cut-off: $r_c = 2.2$)
2. fase liquida: $\rho^\star = 0.8$, $T^\star = 1.1$ (raggio di cut-off: $r_c = 2.5$)
3. fase gassosa: $\rho^\star = 0.05$, $T^\star = 1.2$ (raggio di cut-off: $r_c = 5.0$)

Per concludere, si riportano i risultati in unità del Sistema internazionale delle misure.

In unità LJ, i risultati ottenuti sono equivalenti per entrambi i sistemi. Solo dopo aver riportato il sistema in SI si possono identificare le differenze.

**Fase solida**: per equilibrare il sistema è necessario immergerlo 2/3 volte nel bagno termico. La finestra nella quale sono stati graficati i valori della temperatura contiene una linea rappresentante la temperatura target. Vi è uno scarto molto piccolo tra il valore teorico e quello ottenuto: il sistema si è equilibrato nel modo corretto.

**Fase liquida**: tale fase è meno equilibrata rispetto alla precedente. Si osserva che la temperatura ottenuta differisce di circa 10 K  da quella target. <br>
Si noti che l'energia potenziale per particella è diminuita rispetto a quella della fase solida; questo è dovuto al fatto che l'interazione tra due particelle segue il potenziale di LJ ma fino al raggio di cut-off. Cambiando tale parametro si dà spazio a più o meno interazioni tra le particelle le quali determineranno lo stato fisico del sistema.
Lo stesso ragionamento è valido per la pressione.

**Fase gassosa**: dalle figure si osserva che il gas non raggiunge l'equilibrio. La convergenza potrebbe essere raggiunta utilizzando probabilmente un algoritmo meno approssimato.

```python
print("                                           ARGON SOLIDO")
fig, (f1, f2) = plt.subplots(1,2,figsize=(15,4))

x,y,error = np.loadtxt("./Argon/Solido/output_ekin.dat", usecols=(0,1,2), unpack='true')
f1.errorbar(x,y, yerr=error,label='Energia cinetica (J)',marker=".",color="lightcoral")
leg = f1.legend();
f1.grid(True)

x,y,error = np.loadtxt("./Argon/Solido/output_epot.dat", usecols=(0,1,2), unpack='true')
f2.errorbar(x,y, yerr=error,label='Energia potenziale (J)',marker=".",color="firebrick")
leg = f2.legend()
f2.grid(True)
plt.show()

fig, (f3) = plt.subplots(1,1,figsize=(7,3.5))

x,y,error = np.loadtxt("./Argon/Solido/output_etot.dat", usecols=(0,1,2), unpack='true')
f3.errorbar(x,y, yerr=error,label='Energia totale (J)',marker=".",color="orange")
leg = f3.legend();
f3.grid(True)
plt.show()

fig, (f4, f5) = plt.subplots(1,2,figsize=(15,4))
x,y,error = np.loadtxt("./Argon/Solido/output_temp.dat", usecols=(0,1,2), unpack='true')
f4.errorbar(x,y, yerr=error,label='Temperatura (K)',marker=".",color="coral")
leg = f4.legend();
x=np.linspace(0,100,1000)
y=(x/x)*0.8*120
f4.plot(x,y,color="red")
f4.grid(True)

x,y,error = np.loadtxt("./Argon/Solido/output_press.dat", usecols=(0,1,2), unpack='true')
f5.errorbar(x,y/101300, yerr=error/101300,label='Pressione (Pa)',marker=".",color="tomato")
leg = f5.legend();
f5.grid(True)
plt.show()
```
