Il Supervised Machine Learning prevede che dato un set di dati in input e output, si trovi la migliore funzione che leghi tali dati. A tale scopo si suddividono i dati in due sottoinsiemi: training e test.
I primi sono fondamentali per addestrare il calcolatore, i secondi sono utili per testare la funzione trovata. <br>
Tutto questo è realizzato tramite le reti neurali. <br>
Ciascuna è costituita da diversi livelli caratterizzati da diversi nodi.
Ogni nodo possiede diversi parametri che verranno ottimizzati in modo da minimizzare la funzione costo. 
Quest'ultimo processo è affidato al calcolatore nella fase di training. <br>
Per concludere si chiede di calcolare l'output dei dati in ingresso a partire dai dati test. Poichè viene fornito il risultato esatto, si possono confrontare i risultati.

### Esercizio 1.1.

Una prima applicazione è la regressione lineare per la retta
$$
f(x)=2x+1
$$

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect
```

Vengono generati i dati di training e test sporcando con un errore gaussiano l'output della retta fornita.
La deviazione standard della gaussiana non deve essere troppo alta altrimenti il calcolatore fa fatica ad imparare.

```python
# generate training inputs
np.random.seed(0)

sigma = 0.05 # noise standard deviation, for the moment it is absent

x_train = np.random.uniform(-1, 1, 1000)
y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters

x_valid = np.random.uniform(-1, 1, 10)
x_valid.sort()
y_valid = np.random.normal(m * x_valid + b, sigma)

y_target = m * x_valid + b # ideal (target) linear function

# plot training and target dataset
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True)
plt.show()
```

Un volta introdotti i dati, si costruisce la rete.
In questo caso, la struttura è sequenziale e dotata di un solo livello nascosto.
Creo un solo nodo senza funzione di attivazione poichè l'output di default è **wx+b**, dove **w** e **b** sono i parametri che si vuole minimizzare. Questo suggerisce che per tale problema basta un solo neurone. <br>
Si è scelto come ottimizzatore lo Stochastic gradient descent optimizer mentre la funzione costo è l'errore quadratico medio.

```python
model = tf.keras.Sequential()

model.add(Dense(1, input_shape=(1,)))

model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

model.summary()
```

Adesso inizia il vero e proprio processo di apprendimento. Si dà in pasto alla rete i dati di training che serviranno per trovare il set di parametri che minimizzeranno la funzione costo.

```python
history = model.fit(x=x_train, y=y_train, 
          batch_size=10, epochs=50,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
# return weights
model.get_weights()
```

```python
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

Si nota che all'aumentare del numero delle epoche, la funzione costo diminuisce. Questo è dovuto al fatto che la rete continua ad imparare dai dati perfezionando il set di parametri d'interesse.

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.grid(True)
plt.show()
```

La rete ha imparato ma adesso deve essere interrogata. <br>
Vengono forniti dei dati di input per calcolare dei dati di output.
Si osserva un ragionevole accordo tra i dati ottenuti e quelli ricavati dalla soluzione esatta.

```python
x_predicted = np.random.uniform(-1, 1, 100)
print("Input dataset shape: ", x_predicted.shape)
y_predicted = model.predict(x_predicted)
print("Output dataset shape: ", y_predicted.shape)


plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

### Exercise 11.2

Si vuole estendere il procedimento di cui sopra a funzioni più complesse come

$$
f(x)=4-3x-2x^2+3x^3
$$
per $x \in [-1,1]$.


Provando diverse configurazioni si nota che se si dovesse scegliere un numero troppo basso di neuroni, si otterrebbe un cattivo fit con valori troppo alti della funzione loss. Questo suggerisce di aumentare i livelli e i neuroni fino ad arrivare ad un risultato ragionevole. La struttura finale nascosta è composta da


- Un neurone che riceve un solo valore 
- 25 Neuroni con funzione di attivazione relu
- 45 Neuroni con funzione di attivazione softplus
- 25 Neuroni con funzione di attivazione softplus
- Un neurone senza funzione di attivazione

  
Make graphs comparing fits for different NNs.
 How well do your NN do on points in the range of $x$ where you trained the model? How about points outside the original training data set? 
Summarize what you have learned about the relationship between model complexity (number of parameters), goodness of fit on training data, and the ability to predict well.

```python
# target parameters of f(x) = 4-3x-2x^2+3x^3
a = 3 # slope
b = 2 # intersect
c = 3
d = 4# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 5000)
x_valid = np.random.uniform(-1, 1, 100)
x_valid.sort()
y_target = d-c*x_valid-b*x_valid**2+a*pow(x_valid,3) # ideal (target) linear function

sigma = 0.015 # noise standard deviation, for the moment it is absent
y_train = np.random.normal(d-c*x_train-b*x_train**2+a*pow(x_train,3), sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(d-c*x_valid-b*x_valid**2+a*pow(x_valid,3), sigma)
                           
                          
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True)
plt.show()
```

```python
model = tf.keras.Sequential()

model.add(Dense(1, input_shape=(1,)))

model.add(Dense(25,activation='relu'))

model.add(Dense(45,activation='softplus'))

model.add(Dense(25,activation='softplus'))

model.add(Dense(1,))

model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

model.summary()
```

```python
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=20,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
# Test dei dati rimanenti.
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.grid(True)
plt.show()
```

Si osserva una buona compatibilità con i dati ottenuti dalla funzione target all'interno del range fornito.
Ai bordi sembra che il modello ottenuto non funzioni molto bene; questo è dovuto al fatto che la rete non ha avuto abbastanza punti per imparare.
Si nota che fuori dal range il fit è completamente sbagliato. La rete non può intuire l'andamento della funzione oltre all'intervallo nel quale dove ha appreso.

```python
x_predicted = np.random.uniform(-1.5, 1.5, 100)
print("Input dataset shape: ", x_predicted.shape)
y_predicted = model.predict(x_predicted, batch_size=128)
print("Input dataset shape: ", y_predicted.shape)


plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

### Exercise 11.3
  
Ripetere lo stesso procedimento per la funzione $f(x,y) = \sin(x^2+y^2)$ nel range $x \in [-3/2,3/2]$ and $y \in [-3/2,3/2]$.

```python
np.random.seed(0)
X_train = np.random.uniform(-1.5, 1.5, size=10000)
Y_train = np.random.uniform(-1.5, 1.5, size=10000)

X_valid = np.random.uniform(-1.5, 1.5, size=1000)
Y_valid = np.random.uniform(-1.5, 1.5, size=1000)


Z_target = np.sin(X_valid**2+Y_valid**2) # ideal (target) linear function

sigma = 0.1
Z_train = np.random.normal(np.sin(X_train**2+Y_train**2), sigma) 
Z_valid = np.random.normal(np.sin(X_valid**2+Y_valid**2), sigma)
                           
            

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111 ,projection='3d')
ax.scatter(X_valid, Y_valid, Z_valid, marker='.',color="r")
ax.scatter(X_valid, Y_valid, Z_target, marker='.',color="b")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(20, 30)

plt.show()
```

```python
model = tf.keras.Sequential()

model.add(Dense(1, input_shape=(1,)))

model.add(Dense(25,activation='tanh'))

model.add(Dense(25,activation='tanh'))

model.add(Dense(25,activation='tanh'))

model.add(Dense(1,))

model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

model.summary()
```

```python
lst = []
lst.append(X_train)
lst.append(Y_train)
dati_train = [X_train, Y_train]  # Lista per i dati di training
dati_valid = [X_valid, Y_valid]

history = model.fit(x=dati_train, y=Z_train,
          batch_size=32, epochs=30,
          shuffle=True, 
          validation_data=(dati_valid, Z_valid))
```

```python
score = model.evaluate(dati_valid, Z_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# evaluate model with the exact curve
score = model.evaluate(dati_valid, Z_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.grid(True)
plt.show()
```
